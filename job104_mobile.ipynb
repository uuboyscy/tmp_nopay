```python
'''
以此網頁為例

//www.104.com.tw/job/6pwzi?jobsource=n104bank2
https://www.104.com.tw/job/6pwzi?jobsource=n104bank2
【大數據分析師】
晶鑽生醫股份有限公司
'''

import requests
from bs4 import BeautifulSoup

'''
1. 找出手機版的網址
'''
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.116 Safari/537.36'}
url = 'https://www.104.com.tw/job/6pwzi?jobsource=n104bank2'

web_content_res = requests.get(url, headers=headers)
web_content_soup = BeautifulSoup(web_content_res.text, 'html.parser')

m_content_url = ''
for u in web_content_soup.select('link'):
    try:
        if u['href'].find('m.104.com.tw/job') != -1:
            m_content_url = u['href']
            break
    except KeyError:
        pass
print('手機版網址：', m_content_url)
```

```python
'''
2. 定位工作內容
'''
job_content_res = requests.get(m_content_url, headers=headers)
job_content_soup = BeautifulSoup(job_content_res.text, 'html.parser')
print(job_content_soup.select('div[class="content"]'))
```
